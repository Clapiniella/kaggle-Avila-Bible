```python
scribes = ('Marcus', 'Philippus', 'Franciscus', 'Begonius', 'Paithonius', 'Noaelius', 'Ubuntius', 'FerrumEffractarius')
avila.scribe = pd.Categorical(avila.scribe, ordered = True, categories = scribes)
avila['scribe_c'] = avila.scribe.cat.codes
```

```python
avila.drop(['scribe'], inplace = True, axis=1)
```

```python
avila.head()
```

### Model

```python
X = avila.drop(['scribe_c'], axis= 1) 
y = avila.scribe_c
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
p_test3 = {'learning_rate':[0.15,0.1,0.05,0.01,0.005,0.001], 'n_estimators':[100,250,500,750,1000,1250,1500,1750]}
models = {
    "GradientBoostingClassifier" : GradientBoostingClassifier(),
    "GradientBoostingClassifierP1" : GradientBoostingClassifier(n_estimators=500, learning_rate=0.1),
    "GradientBoostingClassifierP2" : GradientBoostingClassifier(n_estimators=500, learning_rate=0.2),
    "GradientBoostingClassifierP3" : GradientBoostingClassifier(n_estimators=500, learning_rate=0.3)
    #"RandomForestClassifier" : RandomForestClassifier(n_estimators=100),
    #"Logistic" : LogisticRegression(),
    #"SuperVectorC" : SVC()
}

for modelName, model in models.items():
    print(f"Training model: {modelName}")
    model.fit(X_train, y_train)
```

```python
d = {modelName:model.predict(X_test) for modelName, model in models.items()}
df = pd.DataFrame(d)
df["gt"] = y_test.reset_index(drop=True)
df
```

```python
for modelName, model in models.items():    
    print(f"Evaluating model [{modelName}]:")
    print("Accuracy", accuracy_score(df["gt"], df[modelName]))
    print("Precision", precision_score(df["gt"], df[modelName], average='weighted'))
    print("Recall", recall_score(df["gt"], df[modelName], average='weighted'))
```

```python
p_test3 = {'learning_rate':[0.15,0.1,0.05,0.01,0.005,0.001], 'n_estimators':[100,250,500,750,1000,1250,1500,1750]}

tuning = GridSearchCV(estimator =GradientBoostingClassifier(max_depth=4, min_samples_split=2, min_samples_leaf=1, subsample=1,max_features='sqrt', random_state=10), 
            param_grid = p_test3, scoring='accuracy',n_jobs=4,iid=False, cv=5)
tuning.fit(X_train,y_train)
```

```python
d = {'Tuning':tuning.predict(X_test)}
df = pd.DataFrame(d)
df["gt"] = y_test.reset_index(drop=True)
df
```

```python
print("Evaluating model tuning:")
print("Accuracy", accuracy_score(df["gt"], df['Tuning']))
print("Precision", precision_score(df["gt"], df['Tuning'], average='weighted'))
print("Recall", recall_score(df["gt"], df['Tuning'], average='weighted'))
```
